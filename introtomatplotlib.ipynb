```python
ht = pd.read_csv("heart-disease.csv")
ht.head()
```

```python
ht["age"].plot.hist();
```

```python
ht.plot.hist(figsize=(10, 30), subplots=True);
```

```python
ht.head()
```

```python
over = ht[ht["age"] > 50]
over.head()
```

```python
over.plot(kind="scatter", x="age", y="chol", c="target"); #pyplot methord
```

```python
# oo methord

fig ,ax = plt.subplots(figsize=(10, 6))
over.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);

ax.set_xlim([50, 80]);
```

```python
fig, ax = plt.subplots(figsize = [10, 6])

scatter = ax.scatter(x=over['age'], y=over['chol'], c=over['target'])

ax.set(title="Beautiful", xlabel='age', ylabel='chol')

ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over["chol"].mean(), linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)


scatter = ax0.scatter(x=over["age"], y=over["chol"], c=over["target"])
ax0.set(title='Plot1', ylabel='chol')
ax0.axhline(y=over['chol'].mean(), linestyle="--")
ax0.legend(*scatter.legend_elements(), title="Target")

scatter = ax1.scatter(x=over["age"], y=over["thalach"], c=over["target"])
ax1.set(title="Plot2", xlabel="age", ylabel="Thalach")
ax1.axhline(y=over["thalach"].mean(), linestyle="--")
ax1.legend(*scatter.legend_elements(), title="Target");

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```
