```python
import tensorflow as tf
```

```python
from tensorflow import keras 
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train/255
x_test = x_test/255
```

### reshape into 28-28 grid 

```python
import pandas as pd
```

```python
x_train_flatten = x_train.reshape(len(x_train), 28*28)
x_test_flatten = x_test.reshape(len(x_test), 28*28)
```

```python
x_train_flatten[0]
```

### Model building

```python
# We are creating a model which has only one layer i.e., dense layer

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(x_test_flatten, y_test)
```

```python
y_predict = model.predict(x_test_flatten)
```

```python
y_predict
```

```python
plt.matshow(x_test[0]);
```

```python
y_predict[0]
np.argmax(y_predict[0])
```

```python
np.argmax(y_predict[1])
```

```python
plt.matshow(x_test[1]);
```

```python
y_predict_label = [np.argmax(i) for i in y_predict]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predict_label)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth');
```

```python
# Creating a model which has dense layer and a hidden layer

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(x_test_flatten, y_test)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth');
```

```python
# Creating a model which has dense layer, hidden layer and flatten layer

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(100, activation='elu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test, y_test)
```
